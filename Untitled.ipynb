```python
import torchvision

import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
resnet_ = torchvision.models.resnet50(pretrained=True)
weight = resnet_.conv1.weight.clone()
resnet_.conv1 = nn.Conv2d(4,64,kernel_size=7,stride=2,padding=3,bias=False)
resnet_.conv1.weight[:,:3]= weight
nn.init.kaiming_normal_(resnet_.conv1.weight[:,3], nn.init.calculate_gain("relu"))
            
resnet_ = torch.nn.Sequential(*(list(resnet_.children())[:-1]))
print(resnet_)
```
